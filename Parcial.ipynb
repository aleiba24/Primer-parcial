```python
import numpy as np
import statistics as stat

# cargar el archivo
url = '/content/Datos Parcial.csv'
df = pd.read_csv(url)

# Clase 2
comidas = df['COMIDAS DIA']
np.array(comidas)

print(comidas[0])

peso = df['PESO KG']
np.array (peso)

agua = df['VASOS AGUA']
np.array (agua)


print("Suma", peso + agua)
print("Resta", peso - agua)
print("Multiplicacion", peso * agua)
print("Division 1", peso/agua)
print("Division 2", agua/peso)
print("Media", np.mean(peso), np.mean(agua))
print("Desviacion estandar", np.std(peso), np.std(agua))
print("Moda", stat.mode(comidas))

matriz = np.loadtxt(url, delimiter=',', skiprows=1)
print(matriz)

print("Matriz:", matriz)
print("Transpuesta:", matriz.T)
print("Diagonal principal:", np.diag(matriz))
print("Diagonal principal al reves:", np.fliplr(matriz).diagonal())
print("Suma de la matriz", np.sum(matriz))
print("Media:", np.mean(matriz))
print("Mediana:", np.median(matriz))
print("Desviacion estandar:", np.std(matriz))
print("Valor maximo:", np.max(matriz))
print("Valor minimo:", np.min(matriz))
print("Suma de fila:", np.sum(matriz, axis=1))
print("Suma de columna:", np.sum(matriz, axis=0))
matrizToArray = matriz.flatten()
value, count = np.unique(matrizToArray, return_counts=True)
index_mode = np.argmax(count)
mode = value[index_mode]
print("Moda:", mode)
```
